# Q.3 How many records are available in the data set and how many attributes. Do you think the depth (number of records) is sufficient given the breadth? In other words, is the sample likely to be a good representative of the universe?

There are **205** records/observations, **25** predictors/independent variables(X1, X2, X3...X25) and one response/dependent variable(Y) which is 'price'.
The amount of samples you need for a given number of features depends on how the data are distributed, in general, the more features you have, the more data you will need to adequately describe the distribution of the data. This is where some of the rules of thumb come from, such as:

Some rules of thumb:
1. Enough data is needed to provide reliable estimates of the correlations. Use at least **50 cases** and at least **10 to 20** as many cases as there are IVs (as the number of IVs increases, more inferential tests are being conducted (if testing each predictor), therefore more data is needed), otherwise the estimates of the regression line are probably unstable and are unlikely to replicate if the study is repeated.
2. Some follow **15:1 rule** (no. of observations : no. of predictor variables). Here in this dataset, for 25 predictor variables **375** or more observations would be good.
3. **50 + 8(k)** for testing an overall regression model and **104 + k** when testing individual predictors (where k is the number of IVs). Here in this dataset, **50+8(25) = 250** or more observations would be good for testing an overall regression model and **104+25=129** observations or more would be good when testing individual predictors.

These sample size suggestions are based on detecting a medium effect size (β >= .20), with critical α <= .05, with power of 80%.

```python
# Replace the missing value '?' by np.NaN and count the number of NaNs

auto_df[:] = auto_df[:].replace('?', np.NaN)
print('No. of missing values in each feature of the dataset: ')
print(auto_df.isnull().sum())
```

```python
#View of data after replacing '?' by NaN

auto_df.head()
```

```python
# Convert all categorical variables to dummy variables

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

auto_df['make'] = le.fit_transform(auto_df['make'])
auto_df['fuel-type'] = le.fit_transform(auto_df['fuel-type'])
auto_df['aspiration'] = le.fit_transform(auto_df['aspiration'])

# To handle NaNs automatically in 'num-of-doors'
auto_df['num-of-doors'] = auto_df['num-of-doors'].factorize()[0]
auto_df['num-of-doors'] = le.fit_transform(auto_df['num-of-doors'])

auto_df['body-style'] = le.fit_transform(auto_df['body-style'])
auto_df['drive-wheels'] = le.fit_transform(auto_df['drive-wheels'])
auto_df['engine-location'] = le.fit_transform(auto_df['engine-location'])
auto_df['engine-type'] = le.fit_transform(auto_df['engine-type'])
auto_df['num-of-cylinders'] = le.fit_transform(auto_df['num-of-cylinders'])
auto_df['fuel-system'] = le.fit_transform(auto_df['fuel-system'])

auto_df.head()
```

```python
# Missing values imputation

from sklearn.preprocessing import Imputer
imp_mean = Imputer(missing_values='NaN', strategy='mean', axis=0, verbose=0, copy=True)
imp_median = Imputer(missing_values='NaN', strategy='median', axis=0, verbose=0, copy=True)
imp_mode = Imputer(missing_values='NaN', strategy='most_frequent', axis=0, verbose=0, copy=True)

auto_df['normalized-losses'] = imp_median.fit_transform(auto_df['normalized-losses'].values.reshape(-1, 1))
auto_df['num-of-doors'] = imp_mode.fit_transform(auto_df['num-of-doors'].values.reshape(-1, 1))
auto_df['bore'] = imp_mean.fit_transform(auto_df['bore'].values.reshape(-1, 1))
auto_df['stroke'] = imp_mean.fit_transform(auto_df['stroke'].values.reshape(-1, 1))
auto_df['horsepower'] = imp_median.fit_transform(auto_df['horsepower'].values.reshape(-1, 1))
auto_df['peak-rpm'] = imp_median.fit_transform(auto_df['peak-rpm'].values.reshape(-1, 1))
auto_df['price'] = imp_median.fit_transform(auto_df['price'].values.reshape(-1, 1))

auto_df.head()
```

### Splitting data into training and test data

```python
# 80% of the data for training and 20% of the data for testing
from sklearn.model_selection import train_test_split
Y = auto_df['price']
X = auto_df.drop(['price'], axis=1)

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.20, random_state=5)
```

### Feature Selection

```python
from sklearn.feature_selection import SelectPercentile

# 80% of the features/attributes/independent variables/predictors are being selected in the model 
# which are considered as the important predictors of the price of the car

select = SelectPercentile(percentile=80)
select.fit(X_train, Y_train)
X_train_selected = select.transform(X_train)
X_test_selected = select.transform(X_test)

print('X_train.shape ', X_train.shape)
print('X_train_selected.shape ', X_train_selected.shape)

print('X_test.shape ', X_test.shape)
print('X_test_selected.shape ', X_test_selected.shape)

mask = select.get_support()
print(mask)
print()
print('\nSelected features from the dataset for the model are below:\n')
print(X_train.columns[mask])
```

### Training the model

```python
# Create linear regression object
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score

regr = linear_model.LinearRegression()

# Train the model using the training set
regr.fit(X_train_selected, Y_train)
```

```python
# Make predictions using the test set
Y_pred = regr.predict(X_test_selected)

# The coefficients
print('Coefficients: \n', regr.coef_)

# The mean squared error
print('\nMean squared error regression loss = %.2f' % mean_squared_error(Y_test, Y_pred))

# Explained variance score: 1 is perfect prediction
print('\nR^2 coefficient of determination is a statistical measure of how well the regression line approximates the real data points. The closer to 1.0, the better the fit of the regression line. That is, the closer the line passes through all of the points.')
print('\nHere R^2 (coefficient of determination) = explained variance score = %.2f' % r2_score(Y_test, Y_pred))
```

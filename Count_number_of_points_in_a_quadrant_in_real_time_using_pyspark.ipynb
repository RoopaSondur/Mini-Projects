# Count number of points in a quadrant in real time using pyspark

```python
from pyspark import SparkContext, SparkConf
from pyspark.streaming import StreamingContext
conf = SparkConf().setAppName("temp").setMaster("local[3]")
sc = SparkContext(conf=conf)
ssc = StreamingContext(sc, 1)

# global variables
delta = counter_org = counter_1qd = counter_2qd = counter_3qd = counter_4qd = 0

def points_to_quadrants(new_rdd):
    
    # manipulating global variables
    global delta, counter_org, counter_1qd, counter_2qd, counter_3qd, counter_4qd
    
    x = y = delta = 0
    temp = []
    temp = new_rdd.collect()
    print type(temp)
    print 'len(temp) = {}'.format(len(temp))
    print 'temp = {}'.format(temp)
    
    for k in range(0, len(temp)): 
        
        """If messages(points) are waiting in a queue and pyspark program reads all at once, 
        take (x,y) pairwise to seggregate points into 4 quadrants"""
        
        x = int(temp[k + delta])
        y = int(temp[k + delta + 1])
        delta += 1
        
        if (x == 0 and y == 0):
            print 'The point ({},{}) is Origin'.format(x, y)
            counter_org += 1
        elif ((x >= 0 and y > 0) | (x > 0 and y >= 0)):
            print 'The point ({},{}) is in First-quadrant'.format(x, y)
            counter_1qd += 1
        elif (x < 0 and y >= 0):
            print 'The point ({},{}) is in Second-quadrant'.format(x, y)
            counter_2qd += 1
        elif (x < 0 and y < 0): 
            print 'The point ({},{}) is in Third-quadrant'.format(x, y)
            counter_3qd += 1
        elif (x >= 0 and y < 0): 
            print 'The point ({},{}) is in Fourth-quadrant'.format(x, y)
            counter_4qd += 1
        else:            
            print 'Co-ordinates missing'    
            

input_stream = ssc.socketTextStream("20.0.41.53", 7777)
rdd1 = input_stream.map(lambda point: point.replace('(', '')).map(lambda point: point.replace(')', ''))
rdd2 = rdd1.flatMap(lambda point: point.split(","))

# For each RDD in the Discretized Stream (DStream) call the function 'points_to_quadrants(rdd)'
rdd2.foreachRDD(lambda rdd: points_to_quadrants(rdd))


# Start the streaming context and the computation
ssc.start() 

#ssc.awaitTerminationOrTimeout(100)
```

```python
# Stop the streaming context
ssc.stop()

print 'No. of Origin(0,0) points = {}'.format(counter_org)
print 'No. of points in First-quadrant = {}'.format(counter_1qd)
print 'No. of points in Second-quadrant = {}'.format(counter_2qd)
print 'No. of points in Third-quadrant = {}'.format(counter_3qd)
print 'No. of points in Fourth-quadrant = {}'.format(counter_4qd)
```
